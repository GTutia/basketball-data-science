# Player evolution prediction

Goal: try to predict how many points per game that player will have on the next year based on past season data

```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = 100
pd.options.display.max_rows = 500
```

```python
import numpy as np
```

```python
df = pd.read_csv('players_stats_by_season_v2.csv')
```

```python
df.head()
```

## Model 

We are going to use the results from three previous year to try to predict the next year

Let's train using a few variables from the 17/18, 18/19 ,19/20 seasons to predict the 20/21 using a Linear Model

The final test is going to be predicting the 21/22 (without training using any values of 21/22)

```python
df_players = df[df.season.isin(['2020-21','2019-20','2018-19','2017-18'])]
```

```python
season_count = pd.DataFrame(df_players.groupby('name')['season'].count()).reset_index().rename(columns = {'season':'n_seasons'})
```

```python
df_players = df_players[df_players.name.isin(season_count[season_count.n_seasons == 4].name.unique())]
```

```python
df_players.head()
```

```python
df_1718 = df_players[df_players.season == '2017-18'].reset_index(drop = True).set_index('name')
df_1819 = df_players[df_players.season == '2018-19'].reset_index(drop = True).set_index('name')
df_1920 = df_players[df_players.season == '2019-20'].reset_index(drop = True).set_index('name')
df_2021 = df_players[df_players.season == '2020-21'].reset_index(drop = True).set_index('name')
```

```python
y = df_2021.pts_per_g
```

```python
X = df_1920[['height','weight','age','pos','g','gs','mp_per_g','pts_per_g']]
```

```python
X = X.merge(df_1819[['g','gs','mp_per_g','pts_per_g']], left_index = True, right_index = True, suffixes=(None,'_1819'))
X = X.merge(df_1718[['g','gs','mp_per_g','pts_per_g']], left_index = True, right_index = True, suffixes=('_1920','_1718'))
```

```python
X.loc[X.pos.str.contains('PG|SG'),'guard'] = 1
X = X.drop(columns = ['pos'])
```

```python
X = X.fillna(0)
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
X_scaled = pd.DataFrame(X_scaled, columns = X.columns, index = X.index)
```

```python
import statsmodels.api as sm
```

```python
X_scaled = sm.add_constant(X_scaled)
```

```python
results = sm.OLS(y, X_scaled).fit()
```

```python
print(results.summary())
```

## A few conclusions

After running the model, we can notice a few things:
- First, this is a valid model to try to predict the next year points per game, giving the Adjusted R-Squared metric close to 1 (0.768) and the Prob (F-statistic) close to 0 (7.06e-64)
- Among all the variables, only a few of them in fact seems to have a linear relation with the target variable. If we are looking for a confidence interval of 95% the variables that fulfill this criteria (P>|t| < 0.05) are the last year points per game (pts_per_g_1920) and the age. Also, their coefficient signal makes sense with a very usual interpretation:
    - The last year points per game has a positive coefficient (greater than one) and the age has a negative coefficient. Perhaps we could interpretate that as follows: 'A player is in a continuous evolution, it is expected that next season that he has more points per game than in the previous one, but at some point, the age factor starts to be more representative, and then we can start to expect a decrease in the points per game over the years'
- If our criteria of selecting variables would be a bit wider, let's say a 90% confidence interval, other two variables would stand out: the minutes per game of the previous season (mp_per_g_1920) and the total games that he played in the previous season (g_1920). For now, let's keep these variables aside and continue with only the first two for a very simple model 

Let's now remove all the variables that do not have a relation with our target

```python
X2_scaled = X_scaled[['const','pts_per_g_1920','age']]
X2_scaled = X2_scaled.rename(columns = {'pts_per_g_1920':'last_year_pts_per_g'})
X2 = X[['pts_per_g_1920','age']]
X2 = X2.rename(columns = {'pts_per_g_1920':'last_year_pts_per_g'})
```

```python
results2 = sm.OLS(y, X2_scaled).fit()
print(results2.summary())
```

Lets now see how our model would perform in a more classical machine learning framework

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=0.33, random_state=42)
```

```python
scaler2 = StandardScaler()
scaler2.fit(X_train)

X_train_scaled = scaler2.transform(X_train)
X_test_scaled = scaler2.transform(X_test)

X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns, index = X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns, index = X_test.index)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(X_train_scaled, y_train)
```

```python
reg.coef_
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
print(f'The model has scored a RMSE of {mean_squared_error(y_test,reg.predict(X_test_scaled), squared=False)} and a MAE of {mean_absolute_error(y_test,reg.predict(X_test_scaled))}')
```

Now, let's use our model to predict the 21/22 season

```python
df_players = df[df.season.isin(['2020-21','2021-22'])]
season_count = pd.DataFrame(df_players.groupby('name')['season'].count()).reset_index().rename(columns = {'season':'n_seasons'})
df_players = df_players[df_players.name.isin(season_count[season_count.n_seasons == 2].name.unique())]
```

```python
df_2021 = df_players[df_players.season == '2020-21'].reset_index(drop = True).set_index('name')
df_2122 = df_players[df_players.season == '2021-22'].reset_index(drop = True).set_index('name')
```

```python
y_final = df_2122.pts_per_g
```

```python
X_final = df_2021[['pts_per_g','age']]
X_final = X_final.rename(columns = {'pts_per_g':'last_year_pts_per_g'})
```

```python
X_final_scaled = scaler2.transform(X_final)
X_final_scaled = pd.DataFrame(X_final_scaled, columns = X_final.columns, index = X_final.index)
```

```python
print(f'The model has scored a RMSE of {mean_squared_error(y_final,reg.predict(X_final_scaled), squared=False)} and a MAE of {mean_absolute_error(y_final,reg.predict(X_final_scaled))}')
```

We can see now that with a very simple linear model we can predict the next year points per game of a player with a small error, let's look at a few other metrics

```python
df_results = pd.DataFrame(y_final).rename(columns = {'pts_per_g':'pts_per_g_2122'})
```

```python
df_results['predicted_pts_per_g'] = reg.predict(X_final_scaled)
df_results['pts_per_g_2021'] = df_2021[['pts_per_g']]
```

```python
df_results['predicted_increase'] = 1
df_results.loc[df_results.predicted_pts_per_g < df_results.pts_per_g_2021,'predicted_increase'] = 0

df_results['actual_increase'] = 1
df_results.loc[df_results.pts_per_g_2122 < df_results.pts_per_g_2021,'actual_increase'] = 0
```

```python
from sklearn.metrics import confusion_matrix
```

```python
tn, fp, fn, tp = confusion_matrix(df_results.actual_increase,df_results.predicted_increase).ravel()
precision = tp/(tp+fp)
recall = tp/(tp+fn)
```

```python
print(f'We evaluated a total of {df_results.shape[0]} players') 
print(f'\n{tp + fn} of those players increased their point per game from the previous season to the next one.\nThe model has predicted a correct increase in {tp*100.0/(fn+tp):.2f}% of the cases (recall for positive class)')
print(f'\n{fp + tn} of those players decreased their point per game from the previous season to the next one.\nThe model has predicted a correct decrease in {tn*100.0/(fp+tn):.2f}% of the cases (recall for negative class)')
print(f'\nIn this perspective, the model had an overall accuracy of {(tp+tn)*100.0/(tp+fp+tn+fn):.2f}% and a f1_score of {2*(precision+recall)/(precision*recall):.2f}')
```

We can see now, that this very simple model can also be used to predict if a player will improve or decrease its performance from the previous year, of course that this result can be improved with more development

## Final conclusion

Using data from the previous years, it is possible to predict the next year points per game performance. We did it in a very simple way with a simple linear regression and only two variables. Also, this model can be used to try to predict not the exactly points per game but the improvement or not of a player points per game. Of course, that this model can be improved with more development.
